```python
from ultralytics import YOLO

model = YOLO("yolo11s.pt")

# Start hyperparameter tuning using the dict



model.train(
    data="/kaggle/working/NewSonar_dataset2/data.yaml",
    epochs=50,
    imgsz=640,
    batch=-1,
    patience=30,
    save=True,
    save_period=10,
    verbose=True,
    translate=0.1,
    mosaic=0.7,
    fliplr=0.3,
    lr0=0.05,
    lrf=0.9
   
)
```
