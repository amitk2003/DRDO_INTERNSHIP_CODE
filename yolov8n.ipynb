```python
from ultralytics import YOLO

model = YOLO("yolov8n.pt")

model.train(
    data="/kaggle/working/NewSonar_dataset2/data.yaml",
    epochs=50,
    imgsz=640,
    batch=16,
    patience=10,       # Set 0 to disable early stop
    save=True,
    save_period=10,
    verbose=True
   
)
```

```python
from ultralytics import YOLO

# Load pre-trained YOLOv8s model
model = YOLO("/kaggle/working/runs/detect/train/weights/last.pt")  # adjust path if needed

# Train the model with sonar-appropriate hyperparameters
model.train(
    data="/kaggle/working/NewSonar_dataset2/data.yaml",
    epochs=70,
    imgsz=640,
    batch=16,
    patience=0,       # Set 0 to disable early stop
    save=True,
    save_period=10,
    verbose=True,
    translate=0.1,
    scale=0.3,
    fliplr=0.5,
    degrees=5,
    mosaic=0.5,
    lr0=0.01,
    lrf=0.9,
    cache="ram",       # Helps with I/O and speed
    workers=0          # Prevent multiprocessing crashes
)
```
